Extracting Producer Names from 'producers' Column

```python
import ast
pd.set_option('display.max_colwidth', 120)

def extract_value(container,extracted_value):
    bList = []
    if type(container)==type(list()):
        for dictionary in container:
            bList.append(dictionary[extracted_value])
        return bList
    else:
        bList.append(container[extracted_value])
        return bList[0]

def process_row(row,extracted_value="No Need For Extract"):
    try:
        if isinstance(row, str):
            row = ast.literal_eval(row)
        if(extracted_value=="No Need For Extract"):
            return row
        return extract_value(row,extracted_value)
    except:
        return row

df['producers'] = df['producers'].apply(lambda x: process_row(x,"name"))

df["producers"].head(20)
```

Extracting Specific Information from Multiple Columns

```python
columns = ["licensors", "studios", "genres", "themes", "demographics"]
for col in columns:
    df[col] = df[col].apply(lambda x: process_row(x, "name"))

df["broadcast (timezone: Tokyo)"] = df["broadcast (timezone: Tokyo)"].apply(lambda x: process_row(x,"string"))
```

Extracting Start and End Dates from 'aired' Column

```python
from datetime import datetime

def is_valid_date(date_str):
    formats = ["%b %d, %Y", "%Y-%m-%dT%H:%M:%S%z", "%b %Y"]
    
    for fmt in formats:
        try:
            if fmt == "%b %Y":
                date_obj = datetime.strptime(date_str, fmt)
                return date_obj.strftime("%b 1, %Y")
            else:
                date_obj = datetime.strptime(date_str, fmt)
                return date_obj.strftime("%b %d, %Y")
        except ValueError:
            continue
    return None

df["aired"] = df["aired"].apply(lambda x: process_row(x, "string"))
df["aired"] = df["aired"].apply(lambda x: x.split(" to ") if isinstance(x, str) else x)

df["start_airing_date"] = df["aired"].apply(
    lambda x: is_valid_date(x[0]) if (isinstance(x, list) and len(x) > 0) else None
)
df["end_airing_date"] = df["aired"].apply(
    lambda x: is_valid_date(x[1]) if (isinstance(x, list) and len(x) > 1) else None
)

df[["start_airing_date", "end_airing_date"]].head(15)
```

Converting Duration Strings to Minutes

```python
import numpy as np
def convert_to_minutes(duration_str):
    if pd.isna(duration_str):
        return np.nan
    
    duration_str = duration_str.lower()
    minutes = 0
    
    if 'hr' in duration_str:
        hours = int(duration_str.split('hr')[0].strip())
        minutes += hours * 60
        duration_str = duration_str.split('hr')[1].strip()
    
    if 'min' in duration_str:
        mins = int(duration_str.split('min')[0].strip())
        minutes += mins
        duration_str = duration_str.split('min')[1].strip()
    
    if 'sec' in duration_str:
        secs = int(duration_str.split('sec')[0].strip())
        minutes += secs / 60
        
    return minutes

df['duration'] = df['duration'].apply(convert_to_minutes)
```

Filling Missing Duration Values by Anime Type Using Median or Mean Based on Skewness

```python
def fillna_with_stats(x):
    if x.name in ['CM', 'Music', 'ONA', 'OVA', 'PV', 'Special', 'TV', 'TV Special']:
        stat = round(x.median(), 2)
        print(f"Filling missing values in '{x.name}' with median: {stat}")
    else:
        stat = round(x.mean(), 2)
        print(f"Filling missing values in '{x.name}' with mean: {stat}")
    
    return x.fillna(stat)

df['duration'] = df.groupby('type')['duration'].transform(fillna_with_stats)
```

Converting Duration from Minutes to a Human-Readable Format

```python
def minutes_to_duration(minutes):
    if pd.isna(minutes) or minutes <= 0:
        return np.nan
    
    hours = int(minutes // 60)
    mins = int(minutes % 60)
    secs = int((minutes - int(minutes)) * 60)
    
    duration_str = ""
    if hours > 0:
        duration_str += f"{hours} hr "
    if mins > 0:
        duration_str += f"{mins} min "
    if secs > 0:
        duration_str += f"{secs} sec"
    
    return duration_str.strip()

df['duration'] = df['duration'].apply(minutes_to_duration)

df[['type','duration']].head(10)
```

What is the impact of episode count on anime popularity?

```python
df['episodes'] = pd.to_numeric(df['episodes'], errors='coerce')

df_filtered = df.dropna(subset=['episodes']).copy()

bins = [0, 100, 200, 500, 1000, float('inf')]
labels = ['0-100', '100-200', '200-500', '500-1000', '1000+']

df_filtered.loc[:, 'episode_range'] = pd.cut(df_filtered['episodes'], bins=bins, labels=labels, right=False)

popularity_by_range = df_filtered.groupby('episode_range', observed=False)['members'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='episode_range', y='members', data=popularity_by_range, hue='episode_range', palette='viridis', dodge=False)
plt.legend([], [], frameon=False)

plt.title('Impact of Episodes Count on Anime Popularity')
plt.xlabel('Episode Count')
plt.ylabel('Average Number of Members (Popularity)')
plt.show()
```
